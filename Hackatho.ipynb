```python
import numpy as np

# Count frequencies of subgroups
freq_dict = {}
for subgroup in data['Sub_Group_Name']:
    freq_dict[subgroup] = freq_dict.get(subgroup, 0) + 1

# Sort subgroups based on frequencies
sorted_subgroups = sorted(freq_dict.keys(), key=lambda x: freq_dict[x], reverse=True)
sorted_frequencies = [freq_dict[subgroup] for subgroup in sorted_subgroups]

# Create bell curve
x = np.arange(len(sorted_subgroups))
y = sorted_frequencies

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='skyblue', marker='o')
plt.title('Bell Curve of Subgroups')
plt.xlabel('Subgroup')
plt.ylabel('Frequency')
plt.xticks(x, sorted_subgroups, rotation=90)  # Label each x-tick with the subgroup name
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, Dropdown

# Load data from CSV into DataFrame (replace 'data.csv' with your file path)
#data = pd.read_csv('data.csv')

# Count frequencies of subgroups
freq_dict = {}
for subgroup in data['Sub_Group_Name']:
    freq_dict[subgroup] = freq_dict.get(subgroup, 0) + 1

# Sort subgroups based on frequencies
sorted_subgroups = sorted(freq_dict.keys(), key=lambda x: freq_dict[x], reverse=True)
sorted_frequencies = [freq_dict[subgroup] for subgroup in sorted_subgroups]

# Define function to plot bell curve for a specific subgroup
def plot_bell_curve(subgroup):
    subgroup_data = data[data['Sub_Group_Name'] == subgroup]
    subgroup_area_frequencies = subgroup_data['Area_Name'].value_counts()

    plt.figure(figsize=(10, 6))
    subgroup_area_frequencies.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f'Area-wise Analysis for {subgroup}')
    plt.xlabel('Area Name')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Create dropdown widget for selecting subgroup
dropdown = Dropdown(
    options=sorted_subgroups,
    description='Subgroup:',
    disabled=False,
)

# Define function to update plot based on dropdown selection
def update_plot(subgroup):
    plot_bell_curve(subgroup)

# Connect dropdown widget to update_plot function
interact(update_plot, subgroup=dropdown)

# Show initial plot
initial_subgroup = sorted_subgroups[0]
plot_bell_curve(initial_subgroup)
```
